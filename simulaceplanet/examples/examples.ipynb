# Ukázka projektu: Simulace pohybu planet

Tento příklad ukazuje použití balíku `simulaceplanet`, který simuluje gravitační interakce mezi planetami pomocí Newtonových zákonů.


```python
import sys
import os
sys.path.append(os.path.abspath(".."))
```

```python
from simulaceplanet import load_planets_from_json, generate_random_planets, simulate, plot_trajectories, save_animation, create_animation, Planet
```

Načtění planet z .json souboru

```python
planets = load_planets_from_json("planets.json")
```

A nebo vygenerování náhodního počtu planet

```python
planets = generate_random_planets(5)
```

Provedení simulace, simulace se hroutí pokud je časový krok větší než 8.6e4(měsíc)

```python
simulate(planets, dt= 2.628e+3, steps=150000) 
```

Vykreslení trajektorie planet

```python
plot_trajectories(planets)
```

Vytvoření animace pohybu planet a uložení animace do souboru

```python
ani =create_animation(planets, interval=1, kazdy_krok=1000)
save_animation(ani, filename="animation.mp4", fps=30, dpi=300)
```
